```python
from PIL import Image,ImageDraw, ImageFont, ImageFilter, ImageChops
import pandas as pd
import numpy as np
import pathlib
import os
from fontTools.ttLib import TTFont
from fontTools.unicode import Unicode
import cv2 
```

```python
characters = ['1','2','3','4','5','6','7','8','9']
IMG_WIDTH = 100
IMG_HEIGHT = 100
```

```python
def drawCharacter(path, character, font_path):
    background_color = min(int(200 + np.random.normal() * 20),255)
    foreground_color = int(20 + np.random.normal() * 20)
    image = Image.new("L", (IMG_WIDTH, IMG_HEIGHT), (background_color,))
    draw = ImageDraw.Draw(image)
    font = ImageFont.truetype(font_path, 50)
    w, h = draw.textsize(character, font=font)
    draw.text(((IMG_WIDTH - w) / 2, (IMG_HEIGHT - h) / 2), character, (foreground_color), font=font)
    blurred = image.filter(filter=ImageFilter.GaussianBlur(1))

    left = IMG_WIDTH
    top = IMG_HEIGHT
    right = 0
    bottom = 0

    image_pixels = image.load()
    for x in range(image.size[0]):
        for y in range(image.size[1]):    # For every row
            if image_pixels[x,y] != background_color:
                left = min(left, x)
                right = max(right, x)
                top = min(top, y)
                bottom = max(bottom, y)
    
    width = right - left + 4
    height = bottom - top + 4
    x = (left + right)/2
    y = (top + bottom)/2
    
    character_image = blurred.crop((x - width/2, y-height/2, x + width/2, y+height/2))
    pixels = character_image.load() # create the pixel map

    for x in range(character_image.size[0]):
        for y in range(character_image.size[1]):    # For every row
            new_value = pixels[x,y] + np.random.normal() * 2
            pixels[x,y] = (int(new_value),)
            
    character_image=np.array(character_image)  
    
    character_image=cv2.GaussianBlur(character_image,(9,9),0)
    character_image=cv2.adaptiveThreshold(character_image,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
    cv2.THRESH_BINARY,11,2)#

    character_image=cv2.bitwise_not(character_image,character_image)
    
    cv2.imwrite(f'{path}/{count}.png',character_image)
```

```python
def drawblank(path):
    background_color = min(int(200 + np.random.normal() * 20),255)
    foreground_color = int(20 + np.random.normal() * 20)
    image = Image.new("L", (IMG_WIDTH, IMG_HEIGHT), (background_color,))
    blurred = image.filter(filter=ImageFilter.GaussianBlur(1))

    pixels=blurred.load()
    wtop=np.abs(np.random.normal(0,0.01))*blurred.size[0]
    wbottom=blurred.size[0]-np.abs(np.random.normal(0,0.01))*blurred.size[0]
    wleft=np.abs(np.random.normal(0,0.01))*blurred.size[1]
    wright=blurred.size[1]-np.abs(np.random.normal(0,0.01))*blurred.size[1]
    
    for x in range(blurred.size[0]):
        for y in range(blurred.size[1]):# For every row
            if (((x<wtop) or (y<wleft) or (x>wbottom) or (y>wright)) and np.random.choice(2, p=[0.9,0.1])>0):
                new_value=255
                pixels[x,y] = (int(new_value),)
                
    for x in range(blurred.size[0]):
        for y in range(blurred.size[1]):    # For every row
            new_value = pixels[x,y] + np.random.normal() * 2
            pixels[x,y] = (int(new_value),)
    
    
    blurred=cv2.adaptiveThreshold(blurred,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
    cv2.THRESH_BINARY,11,2)#

    blurred=cv2.bitwise_not(blurred,blurred)
    
    cv2.imwrite(f'{path}/{count}.png',blurred)
```

```python
required_ordinals = [ord(glyph) for glyph in ['1','2','3','4','5','6','7','8','9']]

def supportsDigits(fontPath):
    font = TTFont(fontPath)

    for table in font['cmap'].tables:
        for o in required_ordinals:
            if not o in table.cmap.keys():
                return False
    return True
```

```python
count = 0
image_lookup = [];
font_blacklist = [
    '#',
    "italic",
    "Italic",
    "VAZTEK",
    "antique",
    "aztek",
    "RosewoodStd-Regular",
    "ShishoniBrush",
    "VAVOI",
    "seguili",
    'UVNThayGiaoNang_I',
    'VAVOBI',
    'VNI-Nhatban',
    'VNI-Script',
    'VNI-Trung Kien',
    'segoeuii.ttf',
    'VNI-Viettay',
    'Brush',
    'VREDROCK',
    'UVNHaiBaTrung',
    'UVNButLong',
    'AmaticSC',
    'UVNMucCham',
    'UVNThayGiao_BI.TTF',
    'VKUN',
    'segoeuiz.ttf',
    'seguibli.ttf',
    'MyriadPro',
    'Lobster-Regular',
    'Bangers-Regular',
    'VNI',
    'VUSALI.TTF',
    'VPEINOT.TTF',
    'UVNSangSong_R',
    'VDURHAM.TTF',
    'Vnhltfap.ttf',
    'UVNVienDu',
    'UVNBucThu',
    'UVNSangSong',
    'VSCRIPT',
    'VAUCHON',
    'Vnthfap3',
    'VCAMPAI',
    'BungeeOutline',
    'VBROADW',
    'BungeeHairline-Regular',
    'UVNMinhMap',
    'scripti',
    'UVN',
    'brushsbi',
    'Montserrat',
    'VHELVCI.TTF',
    'VFREE',
    'BRUSH'
    'seguis'
    'VSLOGAN.TTF']

def check_blacklist(font):
    for bl in font_blacklist:
        if bl in font:
            return False
    return True

with open('fonts.list', 'r') as fonts:
    for font in fonts:
        font = font.strip()
        can_use = supportsDigits(font)
        if count%100==0:
            print(count)
            
        if can_use and check_blacklist(font):
            for character in characters:
                path = f'testing_data/{character}' if (count % 10) == 0 else f'training_data/{character}'
                image_lookup.append(f'{count}:{font}')
                pathlib.Path(path).mkdir(parents=True, exist_ok=True)
                drawCharacter(path, character, font)
                count = count+1

num=count//9                
for i in range(num):
    path = f'testing_data/0' if (count % 10) == 0 else f'training_data/0'
    pathlib.Path(path).mkdir(parents=True, exist_ok=True)

    drawblank(path)
    count=count+1


with open("image_to_font_map.txt", "w") as map_file:
    map_file.write("\n".join(image_lookup))
```
